# Filter Data Frame

```python
# Filter with boolean logic: Find records in Hawaii
df.loc[df['stateProvince']=='Hawaii'] #brackets are another way to access a column attribute
```

```python
# Filter by list: Find Records in the Pacific

pacific_countries = ["Australia","Fiji","Kiribati","Micronesia","New Zealand","Palau","Papua New Guinea","Samoa","Solomon Islands","Tonga","Vanuatu"]

df_pacific_countries = df.loc[df['country'].isin(pacific_countries)]
df_not_pacific_countries = df.loc[~df['country'].isin(pacific_countries)] #the tilde symbol represents "not"

display('pacific countries',df_pacific_countries,'not pacific countries',df_not_pacific_countries)
```

```python
# Filter by multiple conditions: Find Records in Hawaii or pacific countries
df_filter = df.loc[(df['stateProvince']=='Hawaii')|(df['country'].isin(pacific_countries))] #multiple boolean logic conditions, pipe represents "or"
display(set(df_filter.country),df_filter)
```

```python
# Find Records of Downy Mildew & in Hawaii
df.loc[(df['stateProvince']=='Hawaii')&(df['occurrenceRemarks']=='Downy mildew.')] # the ampersand represents "and"
```

```python
# filter by matching
df.loc[df['sciName'].str.contains('Fusarium')]
```
